# Boston Airbnb Data Analysis

Analyzing the airbnb data for boston and understanding some important business questions...

The dataset contains calendar availability for each of these listings and user reviews on the listings.


## Business Understanding
* Which is the most reviewed and rated places to stay in Boston?
* What is the airbnb price variation in the city and find some of the cheapest places to stay in Boston?
* How Airbnb prices have varied over time in last two year in Boston?

```python
# importing libraries

import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

## Data Understanding

```python
# dataframe input

df_listings = pd.read_csv('listings.csv')
df_calendar = pd.read_csv('calendar.csv')
df_reviews = pd.read_csv('reviews.csv')
```

```python
# view the first entries
df_listings.head()
```

```python
df_calendar.head()
```

```python
df_reviews.head()
```

```python
df_listings.columns
```

## Data Cleaning and Filtering

```python
# removing the null values from important columns 

df_calendar[['price']] = df_calendar[['price']].replace('[\$,]','',regex=True).astype(float)
df_calendar = df_calendar.dropna()

df_listings.price = df_listings.price.str.replace('$', '').str.replace(',', '')
df_listings['price'] = df_listings['price'].astype(float)

df_listings = df_listings.dropna(subset=['review_scores_value'])


df_listings["review_scores_value"].isnull().mean()
df_listings = df_listings[~df_listings["review_scores_value"].isnull()]

```

```python
# filtering data for Boston

df_listings = df_listings[df_listings['city'] == 'Boston']
```

```python
# filtering data needed required for this study, removing extra columns

df_listings_filter = df_listings[['id','host_listings_count',
       'host_total_listings_count', 'neighbourhood_cleansed',
       'city', 'zipcode', 
        'property_type', 'room_type', 'accommodates',
       'bathrooms', 'bedrooms', 'beds', 'bed_type', 'amenities',
       'price', 'number_of_reviews',
       'review_scores_cleanliness',
       'review_scores_checkin', 'review_scores_communication',
       'review_scores_location', 'review_scores_value', 'calculated_host_listings_count',
       'reviews_per_month']]
```

```python
df_listings_filter.head()
```

```python
# replacing index of listings dataframe with id column
# replacing index of calendar dataframe with listing_id column
# replacing index of reviews dataframe with listing_id column

df_listings_filter.set_index('id', inplace=True)
df_calendar.set_index('listing_id', inplace=True)
df_reviews.set_index('listing_id', inplace=True)
```

# Question 1. Most Reviewed and Rated place in Boston

## Data Analysis

```python
# plotting number of reviews vs for different regions

df_listings_filter.groupby('neighbourhood_cleansed').number_of_reviews.sum().plot.bar(figsize =(10,5))
plt.xlabel('Regions');
plt.ylabel('Number of Reviews');
```

Jamaica Plain is highly reviewed...

```python
# total number of reviews for different places in Boston

df_listings_filter.groupby('neighbourhood_cleansed').number_of_reviews.sum()
```

```python
# plotting Average overall rating for all places in Boston

df_listings_filter.groupby('neighbourhood_cleansed').review_scores_value.mean().plot.bar(figsize =(10,5))
plt.xlabel('Region');
plt.ylabel('Net Rating');
```

```python
# average rating for different places in Boston

df_listings_filter.groupby('neighbourhood_cleansed').review_scores_value.mean()
```

Average rating is highest for Leather District

# Question 2. Price Variation over Boston

## Data Preperation

```python
# filtering data needed required for studying price variation, removing extra columns

df_price = df_listings_filter[[ 'neighbourhood_cleansed',
       'city', 'zipcode', 
        'property_type', 'room_type', 'accommodates',
       'bathrooms', 'bedrooms', 'beds', 'bed_type', 'amenities',
       'price']]
```

```python
df_price.head()
```

## Data Analysis

```python
# plotting average price per region

df_price.groupby('neighbourhood_cleansed').price.mean().plot.bar(figsize =(10,5))
plt.xlabel('Region');
plt.ylabel('Price');
```

```python
# plotting price variation on the basis of property type in different regions

plt.figure(figsize=(20,20));
sns.heatmap(df_price.groupby(['neighbourhood_cleansed', 'property_type']).price.mean().unstack(),annot=True, fmt=".0f", cmap="YlOrRd");
```

Cheapest - Rent a villa in Roslindale = $40 / night

```python
# plotting price variation on the basis of type of room in different regions

plt.figure(figsize=(20,20));
sns.heatmap(df_price.groupby(['neighbourhood_cleansed', 'room_type']).price.mean().unstack(),annot=True, fmt=".0f", cmap="YlOrRd");
```

# Question 3. Price variation over time

## Data Preperation

```python
df_calendar.head()
```

```python
# dropping Nan rows from df_calender dataframe

df_calendar = df_calendar.dropna(subset = ['price'])
```

```python
# changing date in df_calendar dataframe from string to datetime type

df_calendar['date'] = pd.to_datetime(df_calendar['date'])
df_calendar
```

```python
# creating new columns for date, year and days in df_calendar dataframe

df_calendar['year'] = df_calendar['date'].dt.year
df_calendar['month'] = df_calendar['date'].dt.month
df_calendar['days'] = df_calendar['date'].dt.dayofweek
```

```python
# adding neighbourhood_cleansed column from df_listings_filter dataframe to df_calendar

df_calendar['neighbourhood_cleansed'] = df_listings_filter['neighbourhood_cleansed']
df_calendar['number_of_reviews'] = df_listings_filter['number_of_reviews']
df_calendar.head()
```

## Data Analysis

```python
# plotting price variation in Boston over months in an year

df_calendar.groupby('month').price.mean().plot.bar(figsize =(10,5))
plt.xlabel('month');
plt.ylabel('Price');
```

September and October are costly!

```python
# average airbnb prices over months in Boston

df_calendar.groupby('month').price.mean()
```

```python
# plotting price variation in different regions of Boston over months in an year


plt.figure(figsize=(20,20));
sns.heatmap(df_calendar.groupby(['neighbourhood_cleansed', 'month']).price.mean().unstack(),annot=True, fmt=".0f", cmap="YlOrRd");
```

```python
# plotting total number of reviews received over months in different regions of Boston


plt.figure(figsize=(20,20));
sns.heatmap(df_calendar.groupby(['neighbourhood_cleansed', 'month']).number_of_reviews.sum().unstack(),annot=True, fmt=".0f", cmap="YlOrRd");
```

Seems to depend more on location than on month. But we generally get higher reviews in December and January. 

```python
# average airbnb prices on different days in a week in Boston


df_calendar.groupby('days').price.mean()
```

Weekdays are costlier overall...
